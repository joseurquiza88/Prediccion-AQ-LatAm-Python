```python
#Funcion para evaluar el desempeño de los ditsintos modelos
def evaluar_modelo(modelo, X_test, y_test):
    """
    Evalúa el desempeño de los modelo de regresión
    """
    pred = modelo.predict(X_test)

    r2 = r2_score(y_test, pred)
    pearson = pearsonr(y_test, pred)[0]
    rmse = np.sqrt(np.mean((pred - y_test) ** 2))
    bias = np.mean(pred - y_test)

    resultados = pd.DataFrame({
        "R2": [round(r2, 5)],
        "Pearson": [round(pearson, 3)],
        "RMSE": [round(rmse, 3)],
        "Bias": [round(bias, 3)],
        "Min_Pred": [round(pred.min(), 3)],
        "Max_Pred": [round(pred.max(), 3)]
    })

    return resultados
```

### Importancia relativa del modelo Random Forest

```python
# -------------- Importancia de variables de RF --------------

#Datos de entreada
estacion = "SP"
modelo = "1"
ruta_modelos = f"D:/Josefina/Proyectos/Tesis/{estacion}/modelos_py/"
tipoModelo = "RF"
modelo_RF_cv = joblib.load(
    ruta_modelos + f"01-{tipoModelo}-CV10-M{modelo}-{estacion}_220126.joblib"
)
print("Caracteristicas del modelo: ",modelo_RF_cv)

#Tomamos la importancia de cada una de las variables y las ordenamos de mayor a menor
importancia_df = pd.DataFrame({
    "Variable": modelo_RF_cv.feature_names_in_,
    "Overall": modelo_RF_cv.feature_importances_
}).sort_values("Overall", ascending=False)
##########################################################
# Plot
plt.figure(figsize=(7,5))
plt.barh(importancia_df["Variable"], importancia_df["Overall"], color=color_SP)
plt.gca().invert_yaxis()
plt.title(f"Importancia de Variables {estacion} para el modelo {tipoModelo}")
plt.xlabel("")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

### Importancia relativa del modelo XGB (gain)

```python
# -------------- Importancia de variables de XGB --------------
#Datos de entreada
estacion = "SP"
modelo = "1"
ruta_modelos = f"D:/Josefina/Proyectos/Tesis/{estacion}/modelos_py/"
tipoModelo = "XGB"
#Levantar el modelo
modelo_XGB_cv = joblib.load(
    ruta_modelos + f"01-{tipoModelo}-CV10-M{modelo}-{estacion}_220126.joblib"
)
print("Caracteristicas del modelo: ",modelo_XGB_cv)


# Importancia por GAIN
importance_dict = modelo_XGB_cv.get_score(importance_type="gain")

importancia_df = (
    pd.DataFrame.from_dict(importance_dict, orient="index", columns=["Overall"])
      .sort_values("Overall", ascending=False)
)
# Nombre de las variables
importancia_df["Variable"] = importancia_df.index

#######################
###Plot
plt.figure(figsize=(7, 5))
plt.barh(importancia_df["Variable"], importancia_df["Overall"], color= color_SP)
plt.gca().invert_yaxis()
plt.title(f"Importancia de Variables {estacion} para el modelo {tipoModelo}")
plt.tight_layout()
plt.show()
```
