```python
# Leer datos de la estacion de interes

# Es incorrecto poruq toma a todas las estaciones
estacion = "SP"
path = (
    f"{estacion}/proceed/merge_tot/{estacion}_merge_comp.csv"
)
data = pd.read_csv(path)
# Setear fotmato fecha
data['date'] = pd.to_datetime(
    data['date'],
    format="%Y-%m-%d"
)
print(data)

# Setear el factor de escala
# Defin los nuevos lmites para PM25 (eje izquierdo) basados en el lmite deseado para AOD (eje derecho)

scale_factor = np.max(data['PM25']) / np.max(data['AOD_055'])

print(scale_factor)


##########################################################################################################

# Plot con serie temporal de ambos dataset PM vs AOD para ver si las tendencias son similares
# Aseguramos formato fecha
data["date"] = pd.to_datetime(data["date"])


fig, ax1 = plt.subplots(figsize=(10, 4))

# --- PM2.5 (eje izquierdo)
ax1.plot(
    data["date"],
    data["PM25"],
    color="#005a32",
    label="PM2.5"
)

ax1.set_ylabel("PM2.5", color="#005a32", fontsize=13)
ax1.tick_params(axis="y", labelcolor="#005a32")
ax1.set_ylim(0, 160)

# --- AOD (eje derecho)
ax2 = ax1.twinx()
ax2.plot(
    data["date"],
    data["AOD_055"] * scale_factor,
    color="#74c476",
    label="AOD"
)

ax2.set_ylabel("AOD", color="#74c476", fontsize=13)
ax2.tick_params(axis="y", labelcolor="#74c476")
ax2.set_ylim(0, 160)

# --- Eje X
ax1.set_xlabel("Fecha", fontsize=13)

# --- Estetica tipo theme_classic
ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)

# --- Titulo estacion
plt.title(f"Evolucin temporal  Estacin {estacion}", fontsize=14)

# --- Sin leyenda
plt.tight_layout()
plt.show()
```

```python
### Prediccion de pm2.5 a partir del AOD
#Primer proxi
estacion = "SP"
modelo = "1"
num_estacion = "04"

base_dir = Path(f"D:/Josefina/Proyectos/ProyectoChile/{estacion}/modelos/ParticionDataSet")

# Dataset de train y test
train_path = base_dir / f"Modelo_{modelo}" / f"M{modelo}_train_{estacion}.csv"
test_path  = base_dir / f"Modelo_{modelo}" / f"M{modelo}_test_{estacion}.csv"

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

print("train_data",train_data.head(2))
print("-------------------------------------------------------------------------------------------------------------------------")
print("-------------------------------------------------------------------------------------------------------------------------")
print("-------------------------------------------------------------------------------------------------------------------------")
print("test_data",test_data.head(2))

###############################################################################################################
### MODELO
# Regresion lineal simple

# Entrenamiento del modelo (PM2.5 ~ AOD)
X_train = train_data[["AOD_055"]]
y_train = train_data["PM25"]

X_test = test_data[["AOD_055"]]
y_test = test_data["PM25"]
#Generacion del modelo 
lm_model = LinearRegression()
# Entrenamiento
lm_model.fit(X_train, y_train)

#Prediccion
predictions_test = lm_model.predict(X_test)
predictions_train = lm_model.predict(X_train)


df_pred = pd.DataFrame({
    "predicciones": predictions_test,
    "observaciones": y_test
})

#Guardar datos
# output_dir = Path("D:/Josefina/Proyectos/Tesis/data/prediccion_RLS")
# output_dir.mkdir(parents=True, exist_ok=True)

# output_path = output_dir / f"{num_estacion}-{estacion}-prediccion_RLS.csv"
# df_pred.to_csv(output_path, index=False)

#Evaluacion del desempeo del modelo
print("Coeficiente (pendiente):", round(lm_model.coef_[0],2))
print("Intercepto:", round(lm_model.intercept_),2)
print("R train:", round(lm_model.score(X_train, y_train),2))
print("R test:", round(lm_model.score(X_test, y_test),2))

#################################################################################
### PLOT
plt.figure(figsize=(5, 5))
plt.scatter(y_test, predictions_test, alpha=0.6,color="#005a32")
plt.plot([0, y_test.max()], [0, y_test.max()], "--", color="black")

plt.xlabel("PM2.5 observado")
plt.ylabel("PM2.5 predicho")
plt.title(f"Regresin lineal PM2.5 ~ AOD\nEstacin {estacion}")

plt.tight_layout()
plt.show()
```
