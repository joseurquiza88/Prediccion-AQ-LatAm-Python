```python
import pandas as pd

# =========================
# Parámetros
# =========================
buffer_time = 120  # minutes
buffer_spatial = "3KM"
city = "SP"
num_estacion = 1

# =========================
# Loop (en R era for(i in 1:1))
# =========================
for i in range(1):  # recorre una vez

    # --- Leer archivos ---
    path_modis = (
        f"D:/Josefina/paper_git/paper_maiac/datasets/V03/processed/"
        f"merge_AER-MODIS/Latam/dia/{buffer_spatial}/{buffer_time}mins/"
        f"{num_estacion}_{city}-{buffer_spatial}-MODIS-{buffer_time}-AER_MEAN.csv"
    )

    path_maiac = (
        f"D:/Josefina/paper_git/paper_maiac/datasets/V03/processed/"
        f"merge_AER-MAIAC/Latam_C61/dia/{buffer_spatial}/{buffer_time}mins/"
        f"{num_estacion}_{city}-{buffer_spatial}-MAIAC-{buffer_time}-AER_MEAN_c61.csv"
    )

    MODIS = pd.read_csv(path_modis)
    MAIAC = pd.read_csv(path_maiac)

    # =========================
    # Formato fecha
    # =========================
    MODIS["date"] = pd.to_datetime(
        MODIS["date"],
        format="%Y-%m-%d %H:%M:%S",
        utc=True
    ).dt.date

    MAIAC["date"] = pd.to_datetime(
        MAIAC["date"],
        format="%Y-%m-%d",
        utc=True
    ).dt.date

    # =========================
    # Merge por fecha
    # =========================
    merge_sat = pd.merge(
        MAIAC,
        MODIS,
        on="date",
        how="inner"
    )

    # =========================
    # Selección de columnas de interés
    # =========================
    merge_sat = pd.DataFrame({
        "date": merge_sat["date"],
        "AOD_modis": merge_sat["AOD_550_modis"],
        "AOD_maiac_61": merge_sat["AOD_550_maiac_mean"],
        "AOD_550_AER_mean": merge_sat["AOD_550_AER_mean_y"]
    })

    print(len(merge_sat))

    # =========================
    # Guardar archivo
    # =========================
    output_path = (
        f"D:/Josefina/paper_git/paper_maiac/datasets/V03/processed/"
        f"merge_AER-MAIAC-MODIS/Latam/dia/{buffer_spatial}/{buffer_time}mins/"
        f"{num_estacion}_{city}-{buffer_spatial}-MAIAC-MODIS-{buffer_time}-AER_MEAN_c61.csv"
    )

    merge_sat.to_csv(output_path, index=False)
```
